```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fre = input().split()
fre = [float(f) for f in fre]
freq = np.array(fre)
```

```python
lp = input().split()
lp = np.array([float(l) for l in lp])
lp
```

```python
hp = input().split()
hp = np.array([float(l) for l in hp])
hp
```

```python
bp = input().split()
bp = np.array([float(l) for l in bp])
bp
```

```python
lpg = lp/3
hpg = hp/3
bpg = bp/3
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d
import math
res =[]
for val in lpg:
    res.append(20*math.log((val),10))
res1 = []
for val in bpg:
    res1.append(20*math.log((val),10))
res2 = []
for val in hpg:
    res2.append(20*math.log((val),10))
plt.plot(fre,res,'o--')
plt.plot(fre,res2,'ro--')
plt.plot(fre,res1,'ko--')

plt.xlabel('Frequency(Hz)')
plt.ylabel('dB')
plt.title('Bode Plot of Transfer function vs Frequency')
plt.legend(['lp','bp','hp'])

plt.savefig('plot')
```
